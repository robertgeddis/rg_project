```python
import pandas as pd

# Convert Bitcoin_Data_Raw.csv into dataframe
file_path = r"c:\Users\rober\Documents\Python\rg_project\data\raw\Bitcoin_Data_Raw.csv"
df_raw_data = pd.read_csv(file_path, index_col='Date', parse_dates=True)

print(df_raw_data.head())
```

```python
df_raw_data.shape 
# (3654, 3)

df_raw_data.dtypes 
# BTC-USD float64, GC=F float64, ^GSPC float64 dtype: object

df_raw_data.isna().sum()
# BTC-USD 0, GC=F 1140, ^GSPC 1140
# Gold + SP500 close weekends and holidays = 1140 NaN

df_raw_data_filled = df_raw_data.ffill().bfill()
# Filling in 1140 with last know price forward (forward + backfill)

df_raw_data_filled.isna().sum()
# BTC-USD 0, GC=F 0, ^GSPC 0
```

```python
# Creating 'clean' version
df_raw_data_filled.to_csv("Bitcoin_Data_Clean.csv")
```
