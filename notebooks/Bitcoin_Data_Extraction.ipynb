```python
%%time
import yfinance as yf
import pandas as pd

# Define yfinance tickers:
## BTC-USD: Bitcoin, GC=F : Gold, ^GSPC: S&P 500
tickers = ["BTC-USD", "GC=F", "^GSPC"]

# Download data - 'threads=True' to help speed up download
data = yf.download(tickers, period="10y", interval="1d", threads=True)

# Using only 'close' prices for simplicity
df_raw_data = data['Close']

print(df_raw_data.tail())

df_raw_data.to_csv("Bitcoin_Data_Raw.csv")
```
